# Excel-ohjelmointi Pythonilla

```python
import pandas as pd
import xlwings as xw
```

### Datan lataaminen Excelistä

Jos data on auki Excelin aktiivisella taulukkovälilehdellä ja datan sisältä on täsmälleen yksi solu valittuna, niin datan lataaminen dataframeen on helppoa:

```python
df = xw.load()
df
```

### Tulokset Exceliin

**xw.view** vie tulokset uuteen Excel-työkirjaan

```python
tunnusluvut = df.describe()
tunnusluvut.index = ['Lukumäärä', 'Keskiarvo', 'Keskihajonta', 'Pienin', 
                     'Alaneljännes', 'Mediaani', 'Yläneljännes', 'Suurin']

xw.view(tunnusluvut, table=False)
```

Python APIn (API=Application Programming Interface) kautta voit ohjelmoida Exceliä. Opettele lukemaan APIn dokumentaatiota: https://docs.xlwings.org/en/stable/api.html

Kaikkia Excelin toimintoja ei ole toteutettu Python APIin, mutta ajan myötä niitä toteutetaan lisää. Puuttuviin toimintoihin pääset Excel APIn kautta.

Värikoodit voit antaa RGB-muodossa (red, green, blue). Värikoodeja löydät esimerkiksi sivulta https://www.rapidtables.com/web/color/RGB_Color.html

Excel APIn vakioita löydät sivulta https://github.com/xlwings/xlwings/blob/main/xlwings/constants.py

```python
# Nimetään Excelin taulukkovälilehti
xw.sheets[0].name = 'Tunnusluvut'

# Asetetaan lukumuotoilu Excelin riveille 3 - 9
xw.sheets('Tunnusluvut').range('3:9').number_format = '0,00'

# Muotoillaan otsikkorivi
rng = xw.sheets('Tunnusluvut').range((1, 1), (1, tunnusluvut.shape[1]+1))
rng.font.bold = True
rng.color = (204, 229, 251)
xw.sheets('Tunnusluvut').used_range.autofit()

# Seuraaviin tarvitaan Excelin APIa

# Tasaus oikeaan reunaan
rng.api.HorizontalAlignment = -4152

# Reunaaviiva alapuolelle
rng.api.Borders(9).Weight = 2

# Tai vaihtoehtoinen pidempi, mutta joidenkin mielestä havainnollisempi tapa:
#rng.api.HorizontalAlignment = xw.constants.HAlign.xlHAlignRight
#rng.api.Borders(xw.constants.BordersIndex.xlEdgeBottom).Weight = xw.constants.BorderWeight.xlThin
```
